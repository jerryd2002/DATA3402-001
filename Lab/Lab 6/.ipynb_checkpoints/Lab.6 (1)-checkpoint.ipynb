```python
def covar_corre_table(dataset):
    df = pd.read_csv(dataset, dtype='float64', names = VarNames)
    data = np.array(df.head())
    covar_matrix = np.cov(data, bias=True)
    corre_matrix = np.corrcoef(data)
    table_1 = covar_matrix
    table_2 = corre_matrix
    
    print("Covariance Matrix")
    display(HTML(tabulate.tabulate(table_1,tablefmt='html', headers = ["A","B","C","D","E"])))

    print("\nCorrelation Matrix")
    display(HTML(tabulate.tabulate(table_2,tablefmt='html', headers = ["A","B","C","D","E"])))

dataset = filename
covar_corre_table(dataset)
```

```python
import matplotlib.pyplot as plt
import numpy as np

tpr = TPR,bins_sig,_=plt.hist(df_sig["M_TR_2"],bins=100, histtype="step",cumulative=-1,density=True)
fpr = FPR,bins_sig,_=plt.hist(df_bkg["M_TR_2"],bins=bins_sig, histtype="step",cumulative=-1,density=True)
```

```python
#5.3

def Plot_Sigi_Obs(df_sig,df_bkg,observations,Numbers_sce,bins = 100, log = False):
    TPR,bins_sig = TPR_FPR(df_sig[observations],bins=bins)
    FPR,bins_sig = TPR_FPR(df_bkg[observations],bins=bins_sig)
    max_sigs=dict()
    table=list()

    for name, (n_sig_expect, n_bkg_expect) in Numbers_sce.items():
        n_sig_expected_prime = n_sig_expect * TPR
        n_bkg_expected_prime = n_bkg_expect * FPR
        sig = n_sig_expected_prime/ np.sqrt(n_sig_expected_prime + n_bkg_expected_prime )
        plt.step(bins_sig[:-1],sig,label=name+" "+str((n_sig_expected, n_bkg_expect)))
        max_i=np.argmax(sig)
        max_sigs[name]=(max_i,n_sig_expected_prime[max_i],n_bkg_expected_prime[max_i],sig[max_i],bins_sig[max_i])
        table.append((name,n_sig_expected, n_bkg_expect, 
                      TPR[max_i],FPR[max_i],
                      n_sig_expected_prime[max_i],n_bkg_expected_prime[max_i],sig[max_i],bins_sig[max_i],max_i)
        )
    if log:
        plt.yscale("log")
    plt.legend()
    plt.show()
    display(HTML(tabulate(table, tablefmt='html',
                                   headers=["Name",'N sig','N bkg',"TPR","FPR","N sig'","N bkg'",'sig','x_c',"bin i"])))
    return max_sigs

Numbers_sce = { "1": (10,100), 
              "2": (100,1000),
              "3": (200,2000),
              "4": (300,3000),
              "5": (400,4000),
              "6": (500,5000),
              "7": (1000,10000),
              "8": (10000,100000)
            }

_= Plot_Sigi_Obs(df_sig,df_bkg,"M_TR_2", Numbers_sce)
```
